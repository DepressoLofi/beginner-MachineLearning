```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

colors = ['cyan', 'yellow', 'pink']
cmp = ListedColormap(colors)

x = [0, 1, 2, 3, 2, 2.5, 2, 3, 1]
y = [0, 0, 0, 1, 2, 3.5, 3, 4, 0.5]
labels = [1, 1, 1, 2, 2, 0, 0, 0, 1]

plt.scatter(x, y, c=labels, cmap = cmp)
plt.show()
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

colors = ['red', 'blue', 'green']  # List of colors
cmap = ListedColormap(colors)  # Create a ListedColormap object

# Usage example: Assign colors to different data points
x = [0, 1, 2, 3]
y = [0, 1, 2, 3]
labels = [0, 1, 0, 2]  # Class labels for each data point

plt.scatter(x, y, c=labels, cmap=cmap)  # Assign colors using the colormap
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))  # Example function

plt.contourf(X, Y, Z, levels=20, cmap='coolwarm')
plt.colorbar()

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Filled Contour Plot')

plt.show()
```
